```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# make bar plot
def plt_bar(df, title, ylim):
    plt.figure(figsize=(8, 6))
    sns.barplot(x=df.index, y='RMSLE', data=df, width=0.5, palette='Blues_d')
    plt.title(title)
    plt.ylim(ylim)
    plt.xlabel('')
    # line at 0.44, red
    plt.axhline(y=0.44, color='r', linestyle='--')
    # gray line at first value of df
    plt.axhline(y=df['RMSLE'][0], color='gray', linestyle='--')
    plt.show()
```

```python
# make DataFrame: [0.46933, 0.45926]
df_1 = pd.DataFrame({'RMSLE': [0.46933, 0.45926]}, index=['C 1-1', 'C 1-2'])
plt_bar(df_1, 'Hypothesis 1', (0.45, 0.47))
```
